```python
data_names = ['tmax', 'tmin', 'srad', 'dl', 'rain', 'snp_v3', "snp_v4"]
data_values = [22.27, 29.64, 9.45, 37.61, 0.0, 0.33, 0.61]

dpi = 80
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
mpl.rcParams.update({'font.size': 10})

#plt.title('Predicate contributions')

ax = plt.axes()
ax.yaxis.grid(True, zorder = 1)

xs = range(len(data_names))

bar = plt.bar([x + 0.005 for x in xs], data_values,
        width = 0.2, color = 'blue', alpha = 0.7,
        zorder = 2)
plt.xticks(xs, data_names)
for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.2lf' % float(height), ha='center', va='bottom')

fig.autofmt_xdate(rotation = 25)
plt.title("Вличиние климатических факторов на модель v4")
plt.legend(loc='upper right')
fig.savefig('bars_autumn.pdf')
```
